```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Given values
x = np.array([[1, 0]])
y = np.array([[1]])

W1 = np.array([[0.2, 0.3],
               [0.4, 0.1]])

W2 = np.array([[0.5],
               [0.6]])

# Forward Propagation
hidden_input = np.dot(x, W1)
hidden_output = sigmoid(hidden_input)

output_input = np.dot(hidden_output, W2)
prediction = sigmoid(output_input)

loss = 0.5 * (y - prediction) ** 2

print("Hidden Layer Output:")
print(hidden_output)

print("\nPrediction:")
print(prediction)

print("\nLoss:")
print(loss)
```

```python
from graphviz import Digraph
from IPython.display import Image, display

dot = Digraph('Network_Architecture', format='png')

# Top to Bottom layout
dot.attr(rankdir='TB')
dot.attr('node', shape='circle', fontsize='12')

# Title
dot.attr(label='Network Architecture', labelloc='t', fontsize='18')

# Input Layer
dot.node('title1', 'Input Layer', shape='plaintext')

dot.node('x1', 'x₁')
dot.node('x2', 'x₂')

# Hidden Layer
dot.node('title2', 'Hidden Layer', shape='plaintext')

dot.node('h1', 'H₁\n(Sigmoid)')
dot.node('h2', 'H₂\n(Sigmoid)')

# Output Layer
dot.node('title3', 'Output Layer', shape='plaintext')

dot.node('o', 'ŷ\n(Sigmoid)')

# Invisible nodes for layer titles
dot.edge('title1', 'x1', style='invis')
dot.edge('title1', 'x2', style='invis')

dot.edge('title2', 'h1', style='invis')
dot.edge('title2', 'h2', style='invis')

dot.edge('title3', 'o', style='invis')

# Connections
dot.edge('x1', 'h1')
dot.edge('x1', 'h2')
dot.edge('x2', 'h1')
dot.edge('x2', 'h2')

dot.edge('h1', 'o')
dot.edge('h2', 'o')

# Save and display
dot.render('network_architecture', cleanup=True)

display(Image(filename='network_architecture.png'))
```
